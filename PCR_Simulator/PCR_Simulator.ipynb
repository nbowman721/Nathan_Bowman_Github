```python
def print_directory_map(dir_path, indent=''):
    """Prints a map of the directory structure."""
    items = os.listdir(dir_path)
    for i, item in enumerate(items):
        item_path = os.path.join(dir_path, item)
        if os.path.isfile(item_path):
            print(indent + '├── ' + item)
        elif os.path.isdir(item_path):
             print(indent + '├── ' + item + '/')
             if i == len(items)-1:
                print_directory_map(item_path, indent + '    ')
             else:
                print_directory_map(item_path, indent + '|   ')

# Print current working directory
current_directory = os.getcwd()
print("Current Directory:", current_directory)

# Print subfolders
items = os.listdir(current_directory)
current_directory_subfolders = [item for item in items if os.path.isdir(os.path.join(current_directory, item))]
print("Subfolders in Current Directory:", current_directory_subfolders)

# Print Directory Map
directory_path = '.'  # Current directory
print(directory_path + '/')
print_directory_map(directory_path)
```

```python
def getReverseComplementNNK(seq):
    complement = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C', 'N': 'N', 'K': 'M', 'M': 'K'}
    return ''.join(complement[base] for base in reversed(seq))

def getPCRProductSequence(plasmid, forward_primer_binding_sequence, reverse_primer_binding_sequence, forward_primer_tail_sequence="", reverse_primer_tail_sequence=""):

    # Reverse primer must match the reverse complement of the plasmid.
    reverse_primer_binding_sequence_rc = getReverseComplementNNK(reverse_primer_binding_sequence)
    reverse_primer_tail_sequence_rc = getReverseComplementNNK(reverse_primer_tail_sequence)

    # Find positions of primers in the circular plasmid.
    plasmid = plasmid.upper()
    forward_pos = plasmid.find(forward_primer_binding_sequence.upper())
    reverse_pos = plasmid.find(reverse_primer_binding_sequence_rc.upper())

    if forward_pos == -1 or reverse_pos == -1:
        return "Primers do not bind to the plasmid."

    # Adjust reverse position to account for the circular nature of the plasmid.
    if reverse_pos < forward_pos:
        reverse_pos += len(plasmid)

    # Extract product sequence without tails.
    product_without_tails = plasmid[forward_pos:reverse_pos] + plasmid[:(reverse_pos % len(plasmid)) + len(reverse_primer_binding_sequence)]

    # Add tails to product sequence
    product_with_tails = forward_primer_tail_sequence + product_without_tails + reverse_primer_tail_sequence_rc

    # make sequence upper case
    product_with_tails_upper = product_with_tails.upper()
    
    return product_with_tails

def getPrimerInformationFromPrimerName(primer_name, primer_dataframe):
    primer_binding_sequence = primer_dataframe.Primer_Binding_Sequence.loc[primer_dataframe.Primer_Name == primer_name].item()
    primer_tail_sequence = primer_dataframe.Primer_Tail_Sequence.loc[primer_dataframe.Primer_Name == primer_name].item()
    primer_sequence = primer_dataframe.Primer_Sequence.loc[primer_dataframe.Primer_Name == primer_name].item()
    primer_description = primer_dataframe.Primer_Description.loc[primer_dataframe.Primer_Name == primer_name].item()
    
    return primer_binding_sequence, primer_tail_sequence, primer_sequence, primer_description

def getPlasmidInformationFromPlasmidName(plasmid_name, plasmid_dataframe):
    plasmid_description = plasmid_dataframe.Plasmid_Description.loc[plasmid_dataframe.Plasmid_Name == plasmid_name].item()
    plasmid_sequence = plasmid_dataframe.Plasmid_Sequence.loc[plasmid_dataframe.Plasmid_Name == plasmid_name].item()
    
    return plasmid_description, plasmid_sequence

def getPCRInformationFromPCRID(pcr_id, pcr_dataframe):
    pcr_set = pcr_dataframe.PCR_Set.loc[pcr_dataframe.PCR_ID == pcr_id].item()
    pcr_num = pcr_dataframe.PCR_Num.loc[pcr_dataframe.PCR_ID == pcr_id].item()
    pcr_description = pcr_dataframe.PCR_Description.loc[pcr_dataframe.PCR_ID == pcr_id].item()
    pcr_template = pcr_dataframe.PCR_Template.loc[pcr_dataframe.PCR_ID == pcr_id].item()
    pcr_forward_primer_name = pcr_dataframe.PCR_Forward_Primer.loc[pcr_dataframe.PCR_ID == pcr_id].item()
    pcr_reverse_primer_name = pcr_dataframe.PCR_Reverse_Primer.loc[pcr_dataframe.PCR_ID == pcr_id].item()
    pcr_product_name = pcr_dataframe.PCR_Product_Name.loc[pcr_dataframe.PCR_ID == pcr_id].item()
    pcr_temperature_profile = pcr_dataframe.PCR_Temperature_Profile.loc[pcr_dataframe.PCR_ID == pcr_id].item()
    
    return pcr_set, pcr_num, pcr_description, pcr_template, pcr_forward_primer_name, pcr_reverse_primer_name, pcr_product_name, pcr_temperature_profile

def getPCRProduct(template_plasmid_name, forward_primer_name, reverse_primer_name, plasmid_dataframe, primer_dataframe):

    # get information about template
    plasmid_description, plasmid_sequence = getPlasmidInformationFromPlasmidName(template_plasmid_name, plasmid_dataframe)
    
    # get information about primers
    forward_primer_binding_sequence, forward_primer_tail_sequence, forward_primer_sequence, forward_primer_description = getPrimerInformationFromPrimerName(forward_primer_name, primer_dataframe)
    reverse_primer_binding_sequence, reverse_primer_tail_sequence, reverse_primer_sequence, reverse_primer_description = getPrimerInformationFromPrimerName(reverse_primer_name, primer_dataframe)

    # get PCR product
    pcr_product_sequence = getPCRProductSequence(plasmid_sequence, forward_primer_binding_sequence, reverse_primer_binding_sequence, forward_primer_tail_sequence, reverse_primer_tail_sequence)

    return pcr_product_sequence

def reindexPCRSequence(pcr_product_sequence, indexing_sequence):
    pcr_product_sequence_reindexLowerSection = pcr_product_sequence[pcr_product_sequence.find(indexing_sequence_p15a):]
    pcr_product_sequence_reindexHigherSection = pcr_product_sequence[0:pcr_product_sequence.find(indexing_sequence_p15a)]
    pcr_product_sequence_reindexed = pcr_product_sequence_reindexLowerSection + pcr_product_sequence_reindexHigherSection

    return pcr_product_sequence_reindexed

# Function to create GenBank file from DNA sequence and annotation DataFrame
def createPlasmidp15aLacIGenbank(sequence, sequence_name, annotations_df):
    #make sequence upper case
    sequence = sequence.upper()
    
    # Create a SeqRecord object for the DNA sequence
    record = SeqRecord(
        Seq(sequence),
        id="DNA1",
        name=sequence_name,
        description="Annotated DNA sequence",
        annotations={"molecule_type": "DNA"}  # Specify the molecule type as DNA
    )

    # Iterate through the DataFrame to add features
    for index, row in annotations_df.iterrows():
        short_seq = row['Annotation_Sequence']
        label = row['Annotation_Name']
        color = row.get('Annotation_Color', None)  # Get the color if specified

        # Find the position of the short sequence in the long sequence
        start = sequence.find(short_seq)
        if start != -1:
            end = start + len(short_seq)
            # Create qualifiers including color if available
            qualifiers = {"label": label}
            if color:
                qualifiers["color"] = color

            # Add a feature for the found sequence
            feature = SeqFeature(FeatureLocation(start, end), type="misc_feature", qualifiers=qualifiers)
            record.features.append(feature)

    # Add CDS of LacI
    LacI_0_7_sequence = "GTGAAACCAGTAACGTTATAC"
    LacI_355_360_sequence = "CGACTGGAAAGCGGGCAGTGA"
    qualifiers = {"label": "lacI" + sequence_name[sequence_name.find("_", 10):]}
    #if color:
        #qualifiers["color"] = color
    start_CDS = sequence.find(LacI_0_7_sequence)
    end_CDS = sequence.find(LacI_355_360_sequence) + len(LacI_355_360_sequence)
    feature = SeqFeature(FeatureLocation(start_CDS, end_CDS), type="CDS", qualifiers=qualifiers)
    record.features.append(feature)
    feature = SeqFeature(FeatureLocation(start_CDS, end_CDS), type="misc_feature", qualifiers=qualifiers)
    record.features.append(feature)

    return record

```

```python
#input template plasmid name and primer names
template_plasmid_name = "p15a_LacI_BbvCI_delLacI75100"
forward_primer_name = "NBO_AD03"
reverse_primer_name = "NBO_AD31"

# get pcr product sequence
pcr_product_sequence = getPCRProduct(template_plasmid_name, forward_primer_name, reverse_primer_name, plasmid_dataframe, primer_dataframe)
#print(len(pcr_product_sequence))
#print(pcr_product_sequence)

# reindex pcr product sequence
indexing_sequence_p15a = "TTTTTTTAAGGCAGTTATTGGTGC"
pcr_product_sequence_reindexed = reindexPCRSequence(pcr_product_sequence, indexing_sequence_p15a)

print(len(pcr_product_sequence_reindexed))
print(pcr_product_sequence_reindexed)
```

```python
# get list of pcr_ids
list_of_pcr_ids = list(pcr_dataframe.PCR_ID)

# iterate through list_of_pcr_ids and get pcr product sequences and append them to list
list_of_pcr_product_sequences = []
for pcr_id in list_of_pcr_ids:
    pcr_set, pcr_num, pcr_description, pcr_template, pcr_forward_primer_name, pcr_reverse_primer_name, pcr_product_name, pcr_temperature_profile = getPCRInformationFromPCRID(pcr_id, pcr_dataframe)
    
    #check if negative control, if not then add pcr_product_sequence to list_of_pcr_product_sequence
    if pcr_forward_primer_name != "no primer" or pcr_reverse_primer_name != "no primer":
        pcr_product_sequence = getPCRProduct(pcr_template, pcr_forward_primer_name, pcr_reverse_primer_name, plasmid_dataframe, primer_dataframe)
    else:
        pcr_product_sequence = ""
    list_of_pcr_product_sequences.append(pcr_product_sequence)

# iterate through list_of_pcr_product_sequences and get reindexed pcr product sequences and append them to list
list_of_reindexed_pcr_product_sequences = []
for pcr_product_sequence in list_of_pcr_product_sequences:
    # reindex pcr product sequence
    indexing_sequence_p15a = "TTTTTTTAAGGCAGTTATTGGTGC"
    reindexed_pcr_product_sequence = reindexPCRSequence(pcr_product_sequence, indexing_sequence_p15a)
    list_of_reindexed_pcr_product_sequences.append(reindexed_pcr_product_sequence)

for i in list_of_reindexed_pcr_product_sequences:
    print(i)
```

```python
#make file folder to receive GenBank Files
subfolder_name = "Output_Files_" + bulk_pcr_set
subfolder_path = os.path.join(current_directory, subfolder_name)
os.makedirs(subfolder_path, exist_ok=True)

#add ligated reindexed pcr product to dataframe
pcr_dataframe["PCR_Product_Ligated_Reindexed_Sequences"] = list_of_reindexed_pcr_product_sequences
pcr_dataframe.to_csv(os.path.join(subfolder_path, pcr_description_notebook_name + "_completed" + ".csv"))
display(pcr_dataframe)

# get GenBank files and add to folder
for row in range(0, len(pcr_dataframe.index)):
    #get row data
    pcr_sample_data = pcr_dataframe.loc[row]

    #check if negative control
    if "T" not in pcr_sample_data.PCR_ID:
        # make GenBank File
        pcr_product_sequence_reindexed = pcr_sample_data.PCR_Product_Ligated_Reindexed_Sequences
        sequence_name = pcr_sample_data.PCR_Product_Name.replace("MNN", "NNK")
        output_name = sequence_name + ".gb"
        record = createPlasmidp15aLacIGenbank(pcr_product_sequence_reindexed, sequence_name, annotation_dataframe)

        # Write the record to a GenBank file
        output_file_name = os.path.join(subfolder_path, output_name)
        with open(output_file_name, "w") as output_handle:
            SeqIO.write(record, output_handle, "genbank")

        print("Wrote " + output_name + " to " + subfolder_path)
```
